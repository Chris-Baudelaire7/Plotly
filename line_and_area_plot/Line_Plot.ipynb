```python
import plotly.express as px 
import plotly.graph_objs as go 
import pandas as pd 
import numpy as np 
from datetime import datetime
import plotly.offline as py

%matplotlib inline 
```

```python
df_stocks = px.data.stocks()
#df_stocks.info()
df_stocks.head()
```

# <span>Simple Line</span>

```python
fig = px.line(df_stocks, x='date', y='GOOG', labels={'x':'Date', 'y':'Price'})
fig.show()
```

# <span>Multi Lines</span>

```python
fig = px.line(df_stocks, x='date', y=['GOOG','AAPL'], 
       labels={'x':'Date', 'y':'Price'}, title='Chronologic Graphics')
py.iplot(go.FigureWidget(fig))
```

### <span style="color:orangered">go.Scatter</span>

##### <span> Multi Line + markers</span>

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AAPL, mode='lines', name='Apple'))
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AMZN, mode='lines+markers', name='Amazon'))
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.GOOG, mode='lines+markers+text', name='Google',
                         line=dict(color='firebrick', width=2, dash='dashdot')))

# ['linear', 'spline', 'hv', 'vh', 'hvh', 'vhv']
fig.update_traces(line=dict(smoothing=1.3, shape="spline"))

fig.update_layout(
    template='plotly_dark',
    plot_bgcolor='#000',
    paper_bgcolor='#000',
    title='Financial Graphics',
    
    xaxis=dict(showline=True, showgrid=True, showticklabels=True,
               linecolor='#555', linewidth=1, ticks='outside',
               tickfont=dict(family='Arial', size=13, color='#555')),
    
    yaxis=dict(showgrid=True, zeroline=False, showline=False, showticklabels=True),
    #autosize=False,
    margin=dict(autoexpand=False, l=0, r=0, t=0, b=0),
    #showlegend=True (par defaut)
    
)
py.iplot(go.FigureWidget(fig))
```

```python
data1 = go.Scatter(x=df_stocks.date, y=df_stocks.AAPL,
                         mode='lines', name='Apple')

data2 = go.Scatter(x=df_stocks.date, y=df_stocks.AMZN,
                         mode='lines+markers', name='Amazon')

data3 = go.Scatter(x=df_stocks.date, y=df_stocks.GOOG,
                         mode='lines+markers',  name='Google',
                         line=dict(color='firebrick', width=2, 
                                   dash='dashdot'))
data = [data1, data2, data3]

layout = go.Layout(title='Financial Graphics',
                   template='plotly_dark',
                   plot_bgcolor='#000',
                   hovermode='y unified',
                   paper_bgcolor='#000',
                   xaxis=dict(
                       showspikes=True,
                        spikedash='dot',
                        spikemode='across',
                        spikesnap='cursor',
                       showline=True, 
                       showgrid=False,
                       showticklabels=True,
                       linecolor='#555',
                       linewidth=2,
                       ticks='outside',
                  tickfont=dict(
                       family='Arial',
                       size=13,
                       color='#555'
                   )
                  ),
                  yaxis=dict(showgrid=False, #zeroline=False, #showline=False,
                             showticklabels=False),
                  #autosize=False,
                  showlegend=False)

fig = go.Figure(data=data, layout=layout)
py.iplot(go.FigureWidget(fig))
```

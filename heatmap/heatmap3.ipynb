```python
def spi():
    path = "/Users/new/Desktop/INFORMATIQUES/_Visualistion_de_données/Plotly/_data/SPEI_41.75_-0.75.csv"
    data = pd.read_csv(path)
    data['DATA'] = pd.to_datetime(data['DATA'], format='%b%Y')
    data['month'] = pd.to_datetime(data['DATA']).dt.month_name()
    data['year'] = pd.to_datetime(data['DATA']).dt.year
    data.drop(columns=["DATA"], inplace=True)
    data = data.reset_index()
    data = data.sort_values(by="index")
    data = data.rename(columns={"index": "id"})
    data = data[["id", "year", "month", "SPEI_48"]]

    fig = go.Figure(
        go.Scatter(
            x=data["id"], y=data["SPEI_48"],
            mode="lines", line=dict(width=1.5), fill="tonexty"
        )
    )

    x = list(data["id"])


    annees_affichees = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
    fig.update_xaxes(tickvals=annees_affichees, ticktext=annees_affichees)



    fig.update_layout(
        template="plotly_dark",
        plot_bgcolor="rgba(0, 0, 0, 0)",
        paper_bgcolor="rgba(0, 0, 0, 0)",
        margin=dict(autoexpand=True, l=0, r=0, b=30),
        font=dict(color="white"),
        hovermode="x",
    )

    return fig

spi()
```

```python
def histogram_global_warm():
    x1 = list(data["SPEI_48"].values)
    
    hist_data = [x1]
    group_labels = ['temperature anomaly'] # name of the dataset

    fig = ff.create_distplot(
        hist_data, group_labels, 
        bin_size=.15, show_rug=False, 
        curve_type='normal', colors=["firebrick"])
    
    fig.add_vline(x=data.SPEI_1.mean(), line=dict(dash="dash", width=.8))
    fig.add_vline(x=data.SPEI_1.std(), line=dict(dash="dash", width=.8))
    fig.add_vline(x=-data.SPEI_1.std(), line=dict(dash="dash", width=.8))


    fig.update_layout(
        bargap=0.01,
        width=700,
        hovermode="x",
        template="plotly_dark",
        plot_bgcolor="rgba(0, 0, 0, 0)",
        paper_bgcolor="rgba(0, 0, 0, 0)",
        margin=dict(autoexpand=True, l=0, r=0, b=30),
        legend=dict(x=0.8, y=.96),
        yaxis=dict(showgrid=False),
        yaxis2=dict(showgrid=False),
        xaxis=dict(showgrid=False),
        barmode='stack'
    )
    
    return fig

histogram_global_warm()
```

### <p style="color:orangered; text-align:center">With scatter charts</p>

```python
import plotly.express as px
import plotly.offline as go
import plotly.graph_objs as go
import pandas as pd
import plotly
```

```python
country_data = px.data.gapminder()
country_data.tail()
```

```python
map_fig = px.scatter_geo(
    country_data, locations="iso_alpha", 
    projection="orthographic", 
    color="continent",
    hover_name="country", 
    hover_data=["lifeExp","pop"])

map_fig.update_layout(template="plotly_dark",
                      margin={"r":0,"t":0,"l":0,"b":0})
map_fig.show()
```

```python
map_fig = px.scatter_geo(
    country_data, locations="iso_alpha", 
    projection="natural earth", 
    color="continent",
    hover_name="country", 
    hover_data=["lifeExp","pop"])

map_fig.show()
```

### <p style="color:orangered; text-align:center">Animation frame</p>

```python
import pandas as pd
import plotly.express as px

us_cities = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv")
us_cities
```

```python
fig = px.scatter_mapbox(
    us_cities, lat="lat", lon="lon", 
    hover_name="City", 
    hover_data=["State", "Population"],
    color_discrete_sequence=["fuchsia"], zoom=3, height=300
)
```

```python
fig.update_layout(
    mapbox_style="white-bg",
    mapbox_layers=[
        {
            "source": [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ]
        }
      ])
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
